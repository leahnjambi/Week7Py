```python
import pandas as pd
```

```python
data = {
    "Name": ["John", "Alice", "Bob", "Eve"],
    "Age": [25, 30, 35, 40],
    "Salary": [50000, 60000, 70000, 80000],
    "Department": ["IT", "HR", "Finance", "IT"]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Save DataFrame to a CSV file
file_path = "data.csv"
df.to_csv(file_path, index=False)
```

```python
print(f"CSV file created: {file_path}")
```

```python
file_path = "data.csv"
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Explore the structure of the dataset
print("\nDataset Information:")
print(df.info())
```

```python
df.dropna(inplace=True)
```

```python
print("\nCleaned Dataset:")
print(df.head())
```

```python
# TASK 2
print("\nBasic Statistics for Numerical Columns:")
print(df.describe())
```

```python
grouped = df.groupby('Department')[['Age', 'Salary']].mean()
print("\nMean Values by Department:")
print(grouped)
```

```python
max_salary_dept = grouped[grouped['Salary'] == grouped['Salary'].max()]
print("\nDepartment with the highest average salary:")
print(max_salary_dept)
```

```python
#TASK 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization 1: Line Chart
plt.figure(figsize=(8, 5))
plt.plot(df["Name"], df["Salary"], marker='o', linestyle='-', color='b')
plt.title("Salary Trend by Name", fontsize=14)
plt.xlabel("Name", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Visualization 2: Bar Chart
plt.figure(figsize=(8, 5))
avg_salary = df.groupby("Department")["Salary"].mean()
avg_salary.plot(kind='bar', color=['teal', 'orange', 'purple'])
plt.title("Average Salary by Department", fontsize=14)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Average Salary", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 3: Histogram
plt.figure(figsize=(8, 5))
plt.hist(df["Age"], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribution of Ages", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# Visualization 4: Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Age", y="Salary", data=df, hue="Department", style="Department", s=100)
plt.title("Relationship Between Age and Salary", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.legend(title="Department")
plt.show()
```
